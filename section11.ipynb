```python
import geopandas as gpd

# خواندن فایل شیپ
gdf = gpd.read_file('/content/landuse_shp.shp')

# بافرهای مورد نظر
buffer_sizes = [100, 200, 300]

# پیمایش تمامی پارسل‌ها
for index, row in gdf.iterrows():
    # شماره و کاربری پارسل مرکزی
    central_parcel_id = row['OBJECTID']
    central_parcel_landuse = row['landuse2']

    print(f"شماره پارسل: {central_parcel_id}, نوع کاربری: {central_parcel_landuse}")

    # پارسل‌های داخل بافرها
    for buffer_size in buffer_sizes:
        buffer = row.geometry.buffer(buffer_size)
        parcels_within_buffer = gdf[gdf.geometry.within(buffer)]

        # تعداد و مساحت پارسل‌ها به تفکیک کاربری
        landuse_counts = parcels_within_buffer['landuse2'].value_counts()
        landuse_areas = parcels_within_buffer.groupby('landuse2')['SHAPE_Area'].sum()

        # محاسبه نسبت کاربری پارسل مرکزی با کاربری‌های دیگر در بافر مورد نظر
        for landuse, count in landuse_counts.items():
            area = landuse_areas.get(landuse, 0)
            total_count = landuse_counts.sum()
            total_area = landuse_areas.sum()

            ratio = (count * area) / (total_count * total_area)
            print(f"   بافر {buffer_size} متر:")
            print(f"      نسبت کاربری {central_parcel_landuse} با کاربری {landuse}: {ratio}")
```
