```python
!pip install gdal
!pip install fiona
!pip install rasterio
!pip install pandas
!pip install geopandas

import geopandas as gpd

# خواندن فایل شیپ
gdf = gpd.read_file('/content/landuse_shp.shp')

# بافرهای مورد نظر
buffer_sizes = [100, 200, 300]

# پیمایش تمامی پارسل‌ها
for index, row in gdf.iterrows():
    # شماره و کاربری پارسل مرکزی
    central_parcel_id = row['OBJECTID']
    central_parcel_landuse = row['landuse2']

    print(f"شماره پارسل: {central_parcel_id}, نوع کاربری: {central_parcel_landuse}")

    # پارسل‌های داخل بافرها
    for buffer_size in buffer_sizes:
        buffer = row.geometry.buffer(buffer_size)
        parcels_within_buffer = gdf[gdf.geometry.within(buffer)]

        # تعداد و مساحت پارسل‌های همنام در منطقه
        similar_parcels = parcels_within_buffer[parcels_within_buffer['landuse2'] == central_parcel_landuse]
        similar_count = len(similar_parcels)
        similar_area = similar_parcels['SHAPE_Area'].sum()

        # تعداد و مساحت کل پارسل‌ها در بافر
        total_count = len(parcels_within_buffer)
        total_area = parcels_within_buffer['SHAPE_Area'].sum()

        # محاسبه نسبت
        ratio = (similar_count * similar_area) / (total_count * total_area)

        print(f"   بافر {buffer_size} متر:")
        print(f"      تعداد پارسل‌های همنام: {similar_count}, مساحت پارسل‌های همنام: {similar_area} متر مربع")
        print(f"      تعداد کل پارسل‌ها در بافر: {total_count}, مساحت کل پارسل‌ها در بافر: {total_area} متر مربع")
        print(f"      نسبت کاربری مرکزی با پارسل‌های همنام: {ratio}")
```
