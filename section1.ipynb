```python
!pip install gdal
!pip install fiona
!pip install rasterio
!pip install pandas
!pip install geopandas


import geopandas as gpd

# مسیر فایل شیپ را تعیین می کنیم
shapefile_path = "/content/landuse_shp.shp"

# خواندن فایل شیپ
gdf = gpd.read_file(shapefile_path)

# تعریف اندازه‌های مختلف بافر
buffer_sizes = [100, 200, 300]

# پایمایش تک تک پارسل‌های شیپ فایل
for index, row in gdf.iterrows():
    parcel_id = row['OBJECTID']  # شماره پارسل
    
    # چاپ شماره پارسل
    print(f"پارسل شماره {parcel_id}:")
    
    # پیدا کردن پارسل‌هایی که کاملاً در داخل بافر 100 متری قرار دارند
    buffer_size = 100
    buffer = row.geometry.buffer(buffer_size)
    parcels_within_buffer = gdf[gdf.geometry.within(buffer)]
    
    # شمارش تعداد پارسل‌ها به تفکیک کاربری2
    landuse_counts = parcels_within_buffer['landuse2'].value_counts().to_dict()
    
    # اضافه کردن کاربری‌هایی که تعداد پارسل‌هایشان صفر است
    all_landuse_types = set(gdf['landuse2'].unique())
    for landuse_type in all_landuse_types:
        landuse_counts.setdefault(landuse_type, 0)
    
    # محاسبه مساحت هر کاربری در بافر 100 متری
    landuse_areas = parcels_within_buffer.groupby('landuse2')['SHAPE_Area'].sum().to_dict()
    
    # چاپ تعداد پارسل‌ها به تفکیک کاربری2
    print(f"تعداد پارسل‌های داخل بافر {buffer_size} متری به تفکیک کاربری2:")
    for landuse_type, count in landuse_counts.items():
        print(f"- {landuse_type}: {count}")
    
    # چاپ مساحت پارسل‌ها به تفکیک کاربری2
    print(f"مساحت پارسل‌ها به تفکیک کاربری2:")
    for landuse_type, area in landuse_areas.items():
        print(f"- {landuse_type}: {area}")
    
    # چاپ تعداد کل پارسل‌ها و مجموع سطح پارسل‌ها در بافر 100 متری
    total_parcels = sum(landuse_counts.values())
    total_area = parcels_within_buffer['SHAPE_Area'].sum()
    print(f"تعداد کل پارسل‌ها: {total_parcels}")
    print(f"مجموع سطح پارسل‌ها: {total_area}")
    print()
    
    # پیدا کردن پارسل‌هایی که کاملاً در داخل بافر 200 متری قرار دارند
    buffer_size = 200
    buffer = row.geometry.buffer(buffer_size)
    parcels_within_buffer = gdf[gdf.geometry.within(buffer)]
    
    # شمارش تعداد پارسل‌ها به تفکیک کاربری2
    landuse_counts = parcels_within_buffer['landuse2'].value_counts().to_dict()
    
    # اضافه کردن کاربری‌هایی که تعداد پارسل‌هایشان صفر است
    all_landuse_types = set(gdf['landuse2'].unique())
    for landuse_type in all_landuse_types:
        landuse_counts.setdefault(landuse_type, 0)
    
    # محاسبه مساحت هر کاربری در بافر 200 متری
    landuse_areas = parcels_within_buffer.groupby('landuse2')['SHAPE_Area'].sum().to_dict()
    
    # چاپ تعداد پارسل‌ها به تفکیک کاربری2
    print(f"تعداد پارسل‌های داخل بافر {buffer_size} متری به تفکیک کاربری2:")
    for landuse_type, count in landuse_counts.items():
        print(f"- {landuse_type}: {count}")
    
    # چاپ مساحت پارسل‌ها به تفکیک کاربری2
    print(f"مساحت پارسل‌ها به تفکیک کاربری2:")
    for landuse_type, area in landuse_areas.items():
        print(f"- {landuse_type}: {area}")
    
    # چاپ تعداد کل پارسل‌ها و مجموع سطح پارسل‌ها در بافر 200 متری
    total_parcels = sum(landuse_counts.values())
    total_area = parcels_within_buffer['SHAPE_Area'].sum()
    print(f"تعداد کل پارسل‌ها: {total_parcels}")
    print(f"مجموع سطح پارسل‌ها: {total_area}")
    print()
    
    # پیدا کردن پارسل‌هایی که کاملاً در داخل بافر 300 متری قرار دارند
    buffer_size = 300
    buffer = row.geometry.buffer(buffer_size)
    parcels_within_buffer = gdf[gdf.geometry.within(buffer)]
    
    # شمارش تعداد پارسل‌ها به تفکیک کاربری2
    landuse_counts = parcels_within_buffer['landuse2'].value_counts().to_dict()
    
    # اضافه کردن کاربری‌هایی که تعداد پارسل‌هایشان صفر است
    all_landuse_types = set(gdf['landuse2'].unique())
    for landuse_type in all_landuse_types:
        landuse_counts.setdefault(landuse_type, 0)
    
    # محاسبه مساحت هر کاربری در بافر 300 متری
    landuse_areas = parcels_within_buffer.groupby('landuse2')['SHAPE_Area'].sum().to_dict()
    
    # چاپ تعداد پارسل‌ها به تفکیک کاربری2
    print(f"تعداد پارسل‌های داخل بافر {buffer_size} متری به تفکیک کاربری2:")
    for landuse_type, count in landuse_counts.items():
        print(f"- {landuse_type}: {count}")
    
    # چاپ مساحت پارسل‌ها به تفکیک کاربری2
    print(f"مساحت پارسل‌ها به تفکیک کاربری2:")
    for landuse_type, area in landuse_areas.items():
        print(f"- {landuse_type}: {area}")
    
    # چاپ تعداد کل پارسل‌ها و مجموع سطح پارسل‌ها در بافر 300 متری
    total_parcels = sum(landuse_counts.values())
    total_area = parcels_within_buffer['SHAPE_Area'].sum()
    print(f"تعداد کل پارسل‌ها: {total_parcels}")
    print(f"مجموع سطح پارسل‌ها: {total_area}")
    print()
```
