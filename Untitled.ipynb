```python
!pip install gdal
!pip install rasterio
!pip install fiona
!pip install pandas
!pip install geopandas

import geopandas as gpd

# مسیر فایل شیپ را تعیین کنید
shapefile_path = "/content/landuse_shp.shp"


# خواندن فایل شیپ
gdf = gpd.read_file(shapefile_path)

# تعریف اندازه‌های مختلف بافر
buffer_sizes = [100, 200, 300]

# لیستی برای ذخیره تعداد پارسل‌های داخل هر بافر
buffer_parcels_count = []

# پایمایش تک تک پارسل‌های شیپ فایل
for index, row in gdf.iterrows():
    parcel_id = row['OBJECTID']  # شماره پارسل
    
    # چاپ شماره پارسل
    print(f"پارسل شماره {parcel_id}:")
    
    # دیکشنری برای ذخیره تعداد پارسل‌های داخل هر بافر به تفکیک کاربری2
    parcel_buffer_counts = {}
    
    # پایمایش تک تک اندازه‌های بافر
    for buffer_size in buffer_sizes:
        buffer = row.geometry.buffer(buffer_size)
        
        # پیدا کردن پارسل‌هایی که کاملاً در داخل هر بافر قرار دارند
        parcels_within_buffer = gdf[gdf.geometry.within(buffer)]
        
        # شمارش تعداد پارسل‌ها به تفکیک کاربری2
        landuse_counts = parcels_within_buffer['landuse2'].value_counts().to_dict()
        
        # مجموع پارسل‌های داخل هر بافر
        total_parcels = sum(landuse_counts.values())
        
        # بررسی کاربری‌هایی که تعداد پارسل صفر دارند
        for landuse_type in gdf['landuse2'].unique():
            if landuse_type not in landuse_counts:
                landuse_counts[landuse_type] = 0
        
        # محاسبه مجموع سطح پارسل‌های داخل هر بافر
        area_sum = parcels_within_buffer['SHAPE_Area'].sum()
        
        # دیکشنری برای ذخیره مساحت هر کاربری2
        landuse_areas = parcels_within_buffer.groupby('landuse2')['SHAPE_Area'].sum().to_dict()
        
        # ذخیره تعداد پارسل‌های داخل هر بافر به تفکیک کاربری2 و مجموع سطح پارسل‌ها و مساحت هر کاربری در دیکشنری
        parcel_buffer_counts[buffer_size] = {"Total": total_parcels, "Total Area": area_sum, **landuse_counts, **landuse_areas}
    
        # چاپ تعداد پارسل‌های داخل هر بافر به تفکیک کاربری2
        print(f"تعداد پارسل‌های داخل بافر {buffer_size} متری به تفکیک کاربری2:")
        for landuse_type, count in landuse_counts.items():
            print(f"- {landuse_type}: {count}")
        print(f"مجموع پارسل‌ها: {total_parcels}")
        print(f"مجموع سطح پارسل‌ها: {area_sum}")
        print()
    
    # چاپ تعداد کل پارسل‌ها
    print(f"تعداد کل پارسل‌ها: {len(gdf)}")
    print()
    
    # چاپ مساحت کل پارسل‌ها
    total_area = gdf['SHAPE_Area'].sum()
    print(f"مساحت کل پارسل‌ها: {total_area}")
    print()

    # چاپ تعداد پارسل‌ها به تفکیک کاربری2 و مساحت هر کاربری
    landuse_counts = gdf['landuse2'].value_counts().to_dict()
    landuse_areas = gdf.groupby('landuse2')['SHAPE_Area'].sum().to_dict()
    print("تعداد پارسل‌ها به تفکیک کاربری2:")
    for landuse_type, count in landuse_counts.items():
        print(f"- {landuse_type}: {count}")
    print()

    print("مساحت پارسل‌ها به تفکیک کاربری2:")
    for landuse_type, area in landuse_areas.items():
        print(f"- {landuse_type}: {area}")
    print()
```
